```python
#importamos librerías y cargamos el archivo para su análisis
import pandas as pd

#la base de datos la guardaremos en una variable llamada empresas
empresas = pd.read_excel("/content/Base-1000-empresas-2023.xlsx")

#visualizamos una parte de los registros para comprobar
#que se cargó correctamente
# en este caso vamos a visualizamos los últimos 7 registros
empresas.tail(7)
```

```python
#vamos a modificar el nombre de una de las columnas de la tabla
#en este ejemplo vamos a cambiar el nombre RANKING por POSICIÓN

empresas = empresas.rename(columns = {'RANKING': 'POSICIÓN'})
#visualizamos si el cambio se efectuó
empresas.columns
```

```python
nuestra_empresa = pd.DataFrame([1001, 748200, 'DESAHUCIADOS', 'LAURA',
                                'Antioquia', 'ANTIOQUIA',
                                'MEDELLIN', 'A0145 - Cría de aves de corral',
                                'AGROPECUARIO', 50000000, -49000000,
                                0, 1000000,  0, 0,0,0, 0, -100, 'Sin Clasificar'])
```

```python
#consultamos las dimensiones de las tablas
print(nuestra_empresa.shape)
print(empresas.shape)
```

```python
#transponemos la información de nuestra_empresa
nuestra_empresa = nuestra_empresa.transpose()
```

```python
#verificamos la nueva dimensión
print(nuestra_empresa.shape)
```

```python
#cambiamos el nombre de las columnas
nuestra_empresa.columns = ['POSICIÓN', 'NIT', 'RAZON SOCIAL', 'SUPERVISOR', 'REGIÓN',
       'DEPARTAMENTO DOMICILIO', 'CIUDAD DOMICILIO', 'CIIU', 'MACROSECTOR',
       'INGRESOS OPERACIONALES 2022', 'GANANCIA (PERDIDA) 2022',
       'TOTAL ACTIVOS 2022', 'TOTAL PASIVOS 2022', 'TOTAL PATRIMONIO 2022',
       'INGRESOS OPERACIONALES 2021', 'GANANCIA (PERDIDA)2021',
       'TOTAL ACTIVOS 2021', 'TOTAL PASIVOS 2021', 'TOTAL PATRIMONIO 2021',
       'GRUPO EN NIIF']
```

```python
nuestra_empresa
```

```python
#concatenamos la información de nuestra_empresa y almacenamos
#la nueva información en empresas
empresas = empresas.append(nuestra_empresa)
```

```python
#guardamos la nueva tabla en un archivo de excel
empresas.to_excel("Nuevas_1000.xlsx", index = False)
```
